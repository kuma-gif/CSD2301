```python
import pandas as pd

# สร้างข้อมูลในรูปแบบ Dictionary
data = {'Name': ['Gong Yoo', 'Hyun Bin', 'Lee Min ho', 'Song Joong Ki', 'Lee Jong suk',
                 'Park Bo-gum', 'Lee Min ho', 'Hyun Bin', 'Lee Dong Wook', 'Lee Minho',
                 'Kim Woo Bin', 'Go Soo'],
        'Age': [26, 24, 23, 22, 23, 24, 26, 24, 22, 23, 24, 24],
        'Score': [85, 63, 55, 74, 31, 77, 85, 63, 42, 62, 89, 77]}

# สร้าง DataFrame จาก Dictionary
df = pd.DataFrame(data)

# แสดง DataFrame
print(df)
```

```python
# สร้างคอลัมน์ใหม่ชื่อ "is_duplicate" ใน DataFrame เพื่อระบุว่าแต่ละแถวเป็นข้อมูลซ้ำหรือไม่ ค่าจะเป็น True ถ้าเป็นข้อมูลซ้ำ และ False ถ้าเป็นข้อมูลที่ไม่ซ้ำ
df["is_duplicate"]= df.duplicated()
# ตัวอย่าง: หากมีข้อมูลซ้ำของคนชื่อ Alice จะมีค่า True ในคอลัมน์ is_duplicate ของแถวที่ซ้ำ
```

```python
# ลบข้อมูลซ้ำทั้งหมดใน DataFrame โดยจะเก็บแถวแรกที่ไม่ซ้ำไว้
df.drop_duplicates()
# ตัวอย่าง: หากมีข้อมูลของ Alice ซ้ำกัน 2 แถว จะเหลือเพียงแถวแรกของ Alice
```

```python
# ลบข้อมูลซ้ำทั้งหมดใน DataFrame โดยจะเก็บแถวสุดท้ายที่ไม่ซ้ำไว้
df.drop_duplicates(keep='last')
# ตัวอย่าง: หากมีข้อมูลของ Alice ซ้ำกัน 2 แถว จะเหลือเพียงแถวสุดท้ายของ Alice
```

```python
# ลบข้อมูลซ้ำทั้งหมดใน DataFrame และเก็บผลลัพธ์ใน DataFrame ใหม่ชื่อ result_df
result_df = df.drop_duplicates(keep=False)
result_df
# ตัวอย่าง: จะได้ DataFrame ที่ไม่มีข้อมูลซ้ำเลย
```

```python
# ลบข้อมูลซ้ำโดยพิจารณาจากคอลัมน์ 'Name' เท่านั้น และเก็บแถวสุดท้ายของแต่ละชื่อไว้
df.drop_duplicates(['Name'], keep='last')
# ตัวอย่าง:หากมีหลายแถวที่มีชื่อ Alice จะเหลือเพียงแถวสุดท้ายของ Alice
```

```python
# สร้างสำเนาของ DataFrame เป็น df1 แล้วลบข้อมูลซ้ำใน df1 โดยไม่สร้าง DataFrame ใหม่
df1 = df.copy( )
```

```python

df1.drop_duplicates(inplace=True)
# ตัวอย่าง:  df1 จะมีข้อมูลที่ไม่ซ้ำกันเท่านั้น
```

```python
# ลบแถวที่มี index เป็น 1 และ 2 ออกจาก DataFrame
df.drop([1,2])
# ตัวอย่าง:  หาก DataFrame มี 5 แถว จะเหลือเพียง 3 แถวที่เหลือ
```

```python
# สร้าง DataFrame ใหม่ที่มีเฉพาะแถวที่คอลัมน์ Name ไม่เท่ากับ 'Gong Yoo'
df[df.Name != 'Gong Yoo ']
# ตัวอย่าง: จะได้ DataFrame ที่ไม่มีข้อมูลของ Gong Yoo
```

```python
# ลบแถวที่มี index เป็น 2 ออกจาก DataFrame
df.drop(df.index[2])
# ตัวอย่าง: เหมือนกับ df.drop([2])
```

```python
# เลือกทุกแถว ยกเว้น 3 แถวสุดท้าย
df[:-3]
# ตัวอย่าง: หาก DataFrame มี 10 แถว จะเหลือเพียง 7 แถวแรก
```

```python
### ตาราง dataFrame
import pandas as pd
import numpy as np  # เพิ่มการนำเข้า numpy

# สร้างข้อมูลในรูปแบบ Dictionary
data = {'Name': ['Gong Yoo', 'Hyun Bin', 'Lee Min ho', 'Song Joong Ki', 'Lee Jong suk',
                 'Park Bo-gum', 'Ji Sung', np.nan],
        'State': ['Arizona', 'Georgia', 'Newyork', 'Indiana', 'Florida',
                  'California', np.nan, np.nan],
        'star': ['Pass', 'Pass', 'Fail', 'Fail', 'Pass',
                 'Pass', np.nan, np.nan],
        'Score': [63.0, 48.0, 56.0, 75.0, np.nan, 77.0, np.nan, np.nan]}

# สร้าง DataFrame จาก Dictionary
df = pd.DataFrame(data)

# แสดง DataFrame
print(df)
```

```python
## (ก) จะลบแถวที่มีค่า NaN ออกจาก DataFrame ทั้งหมด
# คำอธิบาย คำสั่งนี้จะลบแถวทั้งหมดที่มีค่า NaN อย่างน้อย 1 ค่าออกไป
df.dropna()
```

```python
## (ข) คำสั่งนี้เหมือนกับข้อ (ก) คือลบแถวที่มีค่า NaN ออกจาก DataFrame ทั้งหมด
# คำอธิบาย คำสั่งนี้จะเหมือนกับข้อ ก. เนื่องจากการมีค่า NaN อย่างน้อย 1 ค่าในแถว ก็หมายความว่าทั้งแถวนั้นจะถูกพิจารณาว่ามีค่า NaN
df.dropna()
```

```python
## (ค) แสดงคำสั่งใช้สำหรับทำการ drop row ที่มีค่า NaN มากกว่า 2 ค่า
# คำอธิบาย คำสั่งนี้จะลบเฉพาะแถวที่มีค่า NaN มากกว่า 2 ค่าเท่านั้น
df.dropna(thresh=3)
```

```python
## (ง) แสดงคำสั่งใช้สำหรับทำการ drop row ที่มีค่า NaN ใน attribute ‘star’
# คำอธิบาย คำสั่งนี้จะลบเฉพาะแถวที่มีค่า NaN ในคอลัมน์ 'star' เท่านั้น
df.dropna(subset=['star'])
```

```python
## (จ) แสดงคำสั่งใช้สำหรับทำการเติมค่า 0 ให้กับค่า NaN ทุกค่า
# คำอธิบาย คำสั่งนี้จะแทนที่ค่า NaN ทั้งหมดด้วยค่า 0
df.fillna(0)
```

```python
import pandas as pd
import numpy as np

# สร้าง DataFrame ตัวอย่าง
data = {'A': [1, 2, np.nan], 'B': [np.nan, 3, 4], 'C': ['X', 'Y', 'Z']}
df = pd.DataFrame(data)

# เลือกเฉพาะคอลัมน์ตัวเลขและเติมค่า NaN ด้วยค่าเฉลี่ย
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print(df)
```

```python
## (ฉ)  แสดงคำสั่งใช้สำหรับทำการเติมค่าNaN โดยใช้ค่า Mean ของแต่ละ attribute
# คำอธิบาย คำสั่งนี้จะแทนที่ค่า NaN ในแต่ละคอลัมน์ด้วยค่าเฉลี่ยของคอลัมน์นั้น
df.fillna(df.mean())
```

```python
## (ช) แสดงคำสั่งใช้สำหรับทำการเติมค่าNaN โดยใช้ค่า Median ของแต่ละ attribute
# คำอธิบาย คำสั่งนี้จะแทนที่ค่า NaN ในแต่ละคอลัมน์ด้วยค่ามัธยฐานของคอลัมน์นั้น
df.fillna(df.select_dtypes(include=['number']).median()) # Select only numeric columns for median calculation
```
