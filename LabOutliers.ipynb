```python
# .loc เป็นตัวระบุตำแหน่งที่ใช้ในการเข้าถึงข้อมูลใน DataFrame ของ Pandas โดยอ้างอิงตามป้ายกำกับ (labels) ทั้งแถวและคอลัมน์
duplicate_df.loc[:, 'Length'] = duplicate_df['Length'].fillna(duplicate_df['Length'].median()) # แสดงค่าที่ซ้ำกัน
```

```python
duplicate_df.loc[:, 'Height'] = duplicate_df['Height'].fillna(duplicate_df['Height'].median()) # แสดงค่าที่ซ้ำกัน
```

```python
duplicate_df.loc[:, 'Rings'] = duplicate_df['Rings'].fillna(duplicate_df['Rings'].median()) # แสดงค่าที่ซ้ำกัน
```

```python
# ใช้ 3 ตัว[Height ,Whole weight ,Rings]
# [ค่าเฉลี่ย mean ,ค่ามัธยฐาน Median ,ค่าฐานนิยม Rings ]
duplicate_df['Height'] = duplicate_df['Height'].fillna(duplicate_df.groupby('Sex')['Height'].transform('mean'))
duplicate_df['Whole weight'] = duplicate_df['Whole weight'].fillna(duplicate_df.groupby('Sex')['Whole weight'].transform('median'))
duplicate_df['Rings'] = duplicate_df['Rings'].fillna(duplicate_df.groupby('Sex')['Rings'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan))
```

```python
# กำหนดค่าข้อมูลบางคอลัมน์ ต่ำสุด ,มากสุด
lower_bound = 0.0
upper_bound = 1.0

lower_ring = -10.0
upper_ring = 10.0
```

```python
# กรองแถวที่มีค่า 'Whole weight' อยู่ในช่วงที่กำหนด
duplicate_df = duplicate_df[(duplicate_df['Whole weight'] >= lower_bound) & (duplicate_df['Whole weight'] <= upper_bound)]
# กรองแถวที่มีค่า 'Height' อยู่ในช่วงที่กำหนด
duplicate_df = duplicate_df[(duplicate_df['Height'] >= lower_bound) & (duplicate_df['Height'] <= upper_bound)]
# กรองแถวที่มีค่า 'Viscera weight' อยู่ในช่วงที่กำหนด
duplicate_df = duplicate_df[(duplicate_df['Viscera weight'] >= lower_bound) & (duplicate_df['Viscera weight'] <= upper_bound)]

duplicate_df = duplicate_df[(duplicate_df['Rings'] >= lower_ring) & (duplicate_df['Rings'] <= upper_ring)]
```
