```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('generated_transactions.csv')
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.set_index('Item 1', inplace=True)
```

```python
df
```

```python
frequent_itemsets_apriori = apriori(df, min_support=0.3, use_colnames=True)
```

```python
rules_apriori = association_rules(frequent_itemsets_apriori, metric="confidence", min_threshold=0.6)
```

```python
print("Frequent Itemsets from Apriori:\n", frequent_itemsets_apriori)
print("\nAssociation Rules from Apriori:\n", rules_apriori)
```

```python
frequent_itemsets_fpgrowth = fpgrowth(df, min_support=0.3, use_colnames=True)
```

```python
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="confidence", min_threshold=0.6)
```

```python
print("\nFrequent Itemsets from FP-Growth:\n", frequent_itemsets_fpgrowth)
print("\nAssociation Rules from FP-Growth:\n", rules_fpgrowth)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(df).transform(df)
df = pd.DataFrame(te_data, columns=te.columns_)
```

```python
frequent_itemsets = fpgrowth(df, min_support=0.6, use_colnames=True)
```

```python
print(frequent_itemsets)
```
